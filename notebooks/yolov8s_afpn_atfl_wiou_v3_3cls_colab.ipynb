Update `DATA_CFG`, `EPOCHS`, and `BATCH` in the config cell before long training.

```python
from pathlib import Path

REPO_URL = "https://github.com/tuanziya666/yolov8s_kuangjing.git"
REPO_DIR = Path("/content/yolov8s_kuangjing")

if REPO_DIR.exists():
    %cd /content/yolov8s_kuangjing
    !git fetch origin
    !git pull --ff-only origin main
else:
    !git clone {REPO_URL} {REPO_DIR}
    %cd /content/yolov8s_kuangjing

!python -m pip uninstall -y ultralytics >/dev/null 2>&1 || true
!python -m pip install -q -e .
```

```python
from pathlib import Path

# Optional: mount Drive first if your dataset is stored there.
# from google.colab import drive
# drive.mount('/content/drive')

DATA_CFG = "/content/drive/MyDrive/yolo_datasets_10k_yolov8_3class/data.yaml"
MODEL_CFG = "ultralytics/cfg/models/v8/yolov8s_afpn.yaml"
PRETRAINED = "yolov8s.pt"
EPOCHS = 200
IMGSZ = 640
BATCH = 16
WORKERS = 2
DEVICE = 0
PROJECT = "/content/runs/colab"
NAME = "yolov8s_afpn_atfl_wiou_v3_3cls_colab"
SEED = 42
PATIENCE = 100
WIOU_MOMENTUM = 0.01
WIOU_ALPHA = 1.7
WIOU_DELTA = 2.7

if not Path(DATA_CFG).exists():
    print(f"DATA_CFG not found: {DATA_CFG}")
    print("Falling back to coco8 for a smoke test. Replace DATA_CFG with your 3-class dataset before real training.")
    DATA_CFG = "ultralytics/cfg/datasets/coco8.yaml"

assert Path(MODEL_CFG).exists(), f"Missing model config: {MODEL_CFG}"
assert Path("train_yolov8s_afpn_atfl_wiou_v3.py").exists(), "Missing train_yolov8s_afpn_atfl_wiou_v3.py"

print("DATA_CFG =", DATA_CFG)
print("MODEL_CFG =", MODEL_CFG)
print("PRETRAINED =", PRETRAINED)
print("EPOCHS =", EPOCHS)
print("BATCH =", BATCH)
```

```python
!python -u train_yolov8s_afpn_atfl_wiou_v3.py \
  --data {DATA_CFG} \
  --model {MODEL_CFG} \
  --pretrained {PRETRAINED} \
  --epochs {EPOCHS} \
  --batch {BATCH} \
  --imgsz {IMGSZ} \
  --workers {WORKERS} \
  --device {DEVICE} \
  --optimizer SGD \
  --seed {SEED} \
  --patience {PATIENCE} \
  --project {PROJECT} \
  --name {NAME} \
  --wiou-momentum {WIOU_MOMENTUM} \
  --wiou-alpha {WIOU_ALPHA} \
  --wiou-delta {WIOU_DELTA}
```

```python
from pathlib import Path

run_dir = Path(PROJECT) / NAME
print("Run directory:", run_dir)
print("results.csv exists:", (run_dir / "results.csv").exists())
print("best.pt exists:", (run_dir / "weights" / "best.pt").exists())
print("last.pt exists:", (run_dir / "weights" / "last.pt").exists())
```
